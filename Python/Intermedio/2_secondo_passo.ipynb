```python
import pandas as pd

# Creazione del dataframe(dati) = dataset(Tabella)
data = {
    "nome":["Filippo", "Salomon", "Asia", "Catalina", "Camilla", "Ana", "Giuseppe", "Melania", "Alessio", "Camilla", "Cristian","Mattia", "Moussa" ],
    "cognome": ["Battaglia", "Verdi", "Blu", "Birsan", "Noguti", "Bianchi", "Giallini", "Lucio", "Vianello", "Resto", "D'Avena","Torta", "Salisou"],
    "voto":[8,10,10,6,7,8,9,4,7,9,10,6,5]

}

df = pd.DataFrame(data)

df
```

```python
df[df["voto"]>=7]
```

serve a filtrare i dati, in questo caso tutti gli studenti con un voto maggiore o uguale a 7

```python
df["voto"].mean()
```

calcolo la media

```python
df["Promosso"] = df["voto"] > 8
df
```

Aggiorniamo la colonna con Promosso che restituirà True o False in base alla condizione (>8 in questo caso)

```python
df.sort_values(by = "voto", ascending=False)
```

Ordiniamo in maniera discendente (=False)

```python
df.sort_values(by = "voto", ascending=True)
```

Ordiniamo in maniera ascendente (=True) perchè lo è di default (non esiste descending)

```python
df.sort_values(by = "voto", ascending=True, inplace=True)
```

- null = inplace True
- not null = inplace False
- non è obbligatorio inserirlo nel primo
- ma è fondamentale nel secondo
