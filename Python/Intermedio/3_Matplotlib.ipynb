### 2. Grafico a barra raggruppate (Matplotlib)
Utile per:
- confrontare i voti medi di 2 diverse classi di studenti (Classe A e/o Classe B)

```python
import matplotlib.pylab as plt
import numpy as np
# Richiamo le librerie che mi servono

corsi = ['Matematica', 'Italiano', 'Inglese', 'Scienze', 'Educazione Fisica']
voti_classe_A = [9.5, 8.2, 6.8, 7.6, 6.1]
voti_classe_B = [7.5, 6.9, 8.3, 9.0, 6.9]
# Creazione dei dati

x = np.arange(len(corsi)) # imposto la posizione all'etichetta
width = 0.35  # imposta la larghezza della barra

fig, ax = plt.subplots()
rect1 = ax.bar(x -width/2, voti_classe_A, width, label='Classe A')
rect2 = ax.bar(x +width/2, voti_classe_B, width, label='Classe B')

plt.show()
```

```python
import matplotlib.pylab as plt
import numpy as np
# Richiamo le librerie che mi servono

corsi = ['Matematica', 'Italiano', 'Inglese', 'Scienze', 'Informatica']
voti_classe_A = [9.5, 8.2, 6.8, 7.6, 8]
voti_classe_B = [7.5, 6.9, 8.3, 9.0, 7]
# Creazione dei dati

x = np.arange(len(corsi)) # imposto la posizione all'etichetta
# np.arange è la media 
width = 0.35  # imposta la larghezza della barra

fig, ax = plt.subplots() # confronta le classi
rect1 = ax.bar(x -width/2, voti_classe_A, width, label='Classe A')
# scelgo la forma della barra (rect)
# x è la posizione dell'origine della barra (x con + o -) decido dove mettere la barra 
# metto width/2 perche voglio dividere la larghezza della per 2
# decido cosa deve rappresentare (quale delle 2 classi A o B)
# decido come deve essere chiamata la barra 
rect2 = ax.bar(x +width/2, voti_classe_B, width, label='Classe B')

# dall'asse le barre partono, hanno larghezza 0.35

ax.set_title('Confronto Media tra Classi')
ax.set_ylabel('Voto Medio')
ax.set_xticks(x)
ax.set_xticklabels(corsi)
ax.legend()

plt.show()
```

### 3. Grafico Heatmap (Seaborn)
Utile:
- per visualizzare la presenza degli studenti durante le settimane
- densità popolazione
- temperatura


```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
# Richiamo le librerie che mi servono per la resa grafica e di codice

# Dati: righe = studenti, colonne = giorni della settimana (con 1 = Presenza/Presente, 0 = Assenza/Assente)
data = {
    'Lun': [1,1,0,1,1,1,0,1],
    'Mar': [0,1,1,0,1,0,1,1],
    'Mer': [1,1,0,1,0,1,1,1],
    'Gio': [0,0,1,0,1,1,0,0],
    'Ven': [1,1,0,0,1,1,0,0]
}

studenti = ['Giorgio', 'Ana', 'Catalina','Eli', 'Emma', 'Massimo', 'Alessio','Mattia']

df = pd.DataFrame(data, index=studenti)
# dichiaro quale sia il dataframe, indicizzo per studente (facoltativo, potrei anche solo scrivere 'studenti')
sns.heatmap(df, annot=True, cmap='YlGnBu')
# dichiaro quale libreria uso
# quale tipologia di grafico voglio
# cosa mi deve rappresentare (df in questo caso)
# con annot=True scrivo all'interno delle celle (in questo caso 1 e 0)
# se fosse annot=False lascerebbe le celle vuote ma colorate rispetto all'indicazione
# devo inserire il codice colore in questo caso YlGnBu preceduto da cmap (colour of the map)

plt.show()
```

### 4. Grafico a Torta (Matplotlib)

```python
import matplotlib.pyplot as plt

etichette = ['Studenti Lavoratori','Studenti Full Time', 'Uditori', 'Studenti Laureati' ]
valori = [10, 25, 5,2]
plt.title('Tipologia di studenti nel corso')
plt.pie(valori, labels=etichette, autopct='%1.1f%%', startangle=140)
# informazione grafica di grafico a torta 'pie'
# autopct vuol dire che inserisce la percentuale in automatico
# segnalo che voglio un solo decimale dopo la virgola con '%1.1f%%' 
# se ne volessi 2 dovrei cambiare il secondo 1 con il 2

plt.show
```

```python
import matplotlib.pyplot as plt

etichette = ['Studenti Lavoratori','Studenti Full Time', 'Uditori', 'Studenti Laureati' ]
valori = [10, 25, 5,2]
plt.title('Tipologia di studenti nel corso')
plt.pie(valori, labels=etichette, autopct='%1.1f%%', startangle=140)
# informazione grafica di grafico a torta 'pie'
# autopct vuol dire che inserisce la percentuale in automatico
# segnalo che voglio un solo decimale dopo la virgola con '%1.1f%%' 
# se ne volessi 2 dovrei cambiare il secondo 1 con il 2
# startangle specifica l'angolo, espresso in gradi, da cui inizia il disegno del primo spicchio
# ruotato in senso antiorario rispetto all'asse delle x positive.
# posso inserire quanti valori voglio, l'importante è che il numero combaci con il numero di etichette

plt.show
```

### 5.Grafico Radar (Matplotlib + Numpy)
utile per:
- valutare le competenze di un signolo studente in diverse aree

```python
import matplotlib.pyplot as plt
import numpy as np

# Creiamo la lista delle competenze
labels = np.array(['Coding', 'Teoria', 'Progetto', 'Soft Skills','Data Analysis','Presenza','Reverse Engineering'])
stats = np.array([6,8,10,11,6,7,4])

angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False).tolist()
# divide la circonferenza (0, 2*np.pi) in parti uguali rispetto al numero di labels (len(labels))
## in altre parole crea un array di angoli equidistanti tra 0 e 2*pi, con un numero di elementi pari al numero di etichette nelle competenze
# individuo in endpoint con false perchè dico al programma di non finire esattamente a 360°
## poichè 360° coincide con 0°
###ossia esclude il punto finale 2*pi dall'array di angoli, in modo che le etichette siano posizionate in modo circolare senza sovrapporsi
# .tolist converte l'array di angoli in una lista, che può essere utilizzata per posizionare le etichette in un grafico radar o a ragnatela


stats = np.concatenate((stats, [stats[0]]))
# questa indicazione fa sì che si chiuda il cerchio
# aggiunge il primo valore di competenza alla fine dell'array di valori per chiudere il grafico radar
angle = np.concatenate((angles, [stats[0]]))
# questa indicazione fa sì che si chiuda il cerchio
# aggiunge il primo angolo alla fine dell'array di angoli per chiudere il grafico radar

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True)) 
# crea una figura di dimensioni 8x8 pollici e un subplot con coordinate polari per il grafico radar

ax.plot(stats, color='red', linewidth=2, linestyle='solid') 
# traccia una linea che collega i punti del grafico radar, con colore rosso, spessore di 2 e stile solido

ax.fill(stats, color='red', alpha=0.20) 
# riempie l'area sotto la linea del grafico radar con colore rosso e trasparenza del 20%

ax.set_xticks(angles[:-1]) 
# imposta le posizioni delle etichette sui punti del grafico radar
## escludendo l'ultimo punto che è una ripetizione del primo

ax.set_xticklabels(labels) 
# imposta le etichette delle competenze sui punti del grafico radar

ax.set_title('Competenze dello Studente') 
# aggiunge un titolo al grafico radar

plt.show
```
