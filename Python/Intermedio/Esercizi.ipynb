### Creare un dataframe con 4 colonne
<ul> 4 colonne
    <l>nomeCorso</l>
    <l>descrizione</l>
    <l>dataInizio</l>
    <l>dataFine</l>
</ul>

definisco il nome del dataframe (corso) poi definisco subito che lavoro su pandas (pd.) e che andrò a creare un dataframe(.DataFrame)

```python
import pandas as pd

corsi = pd.DataFrame({
    "nomeCorso": ["Python", "SQL", "Excel","Power Query"],
    "descrizione": ["Base", "Intermedio", "Avanzato", "Base"],
    "dataInizio": ["2026-04-01", "2026-05-01", "2025-12-01", "2025-11-01"],
    "dataFine": ["2026-06-01", "2026-07-01", "2026-02-01", "2026-01-01"]
})

corso
```

### Creare un dataframe con 5 colonne
<ul> 4 colonne
    <l>nome</l>
    <l>cognome</l>
    <l>dataNascita</l>
    <l>email</l>
    <l>telefono</l>
</ul>

```python
import pandas as pd

studenti = pd.DataFrame({
    "nome": ["Michael", "Scott", "Tracy","Becky"],
    "cognome": ["Blu", "Rossi", "Verdi", "Bianchi"],
    "dataNascita": ["1992-06-25", "1993-07-27", "1994-08-28", "1995-09-29"],
    "email": ["mb@email.com", "sr@email.com", "tv@email.com", "bb@email.com"],
    "telefono": ["3456789000", "34576589900", "33785984783", "33467589939"]
})

studenti
```

PER SALVARE: to_csv è un metodo o una funzione che permette di salvare un dataframe in un file con formato .csv
index=false indica che non voglio stampare l'indice della lista

```python
studenti.to_csv("studenti.csv", index=False) 
corsi.to_csv("corsi.csv", index=False)
```
