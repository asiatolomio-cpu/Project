### Grafico a barra raggruppate (Matpltlib)
Utile per:
- confrontare i voti medi di 2 diverse classi di studenti (Classe A e/o Classe B)

```python
import matplotlib.pylab as plt
import numpy as np
# Richiamo le librerie che mi servono

corsi = ['Matematica', 'Italiano', 'Inglese', 'Scienze', 'Educazione Fisica']
voti_classe_A = [9.5, 8.2, 6.8, 7.6, 6.1]
voti_classe_B = [7.5, 6.9, 8.3, 9.0, 6.9]
# Creazione dei dati

x = np.arange(len(corsi)) # imposto la posizione all'etichetta
width = 0.35  # imposta la larghezza della barra

fig, ax = plt.subplots()
rect1 = ax.bar(x -width/2, voti_classe_A, width, label='Classe A')
rect2 = ax.bar(x +width/2, voti_classe_B, width, label='Classe B')

plt.show()
```

```python
import matplotlib.pylab as plt
import numpy as np
# Richiamo le librerie che mi servono

corsi = ['Matematica', 'Italiano', 'Inglese', 'Scienze']
voti_classe_A = [9.5, 8.2, 6.8, 7.6]
voti_classe_B = [7.5, 6.9, 8.3, 9.0]
# Creazione dei dati

x = np.arange(len(corsi)) # imposto la posizione all'etichetta
# np.arange è la media 
width = 0.35  # imposta la larghezza della barra

fig, ax = plt.subplots()
rect1 = ax.bar(x -width/2, voti_classe_A, width, label='Classe A')
rect2 = ax.bar(x +width/2, voti_classe_B, width, label='Classe B')
# scelgo la forma della barra, la larghezza e come devono essere accoppiate
# dall'asse le barre partono, hanno larghezza 0.35

ax.set_title('Confronto Media tra Classi')
ax.set_ylabel('Voto Medio')
ax.set_xticklabels(corsi)
ax.legend()

plt.show()
```

Heatmap (Seaborn)
Utile:
- per visualizzare la presenza degli studenti durante le settimane
- densità popolazione
- temperatura


```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
# Richiamo le librerie che mi servono per la resa grafica e di codice

# Dati: righe = studenti, colonne = giorni della settimana (con 1 = Presenza/Presente, 0 = Assenza/Assente)
data ={
    'Lun': [1,1,0,1],
    'Mar': [0,1,1,0],
    'Mer': [1,1,0,1],
    'Gio': [0,0,1,0],
    'Ven': [1,1,0,0]
}

studenti = ['Giorgio', 'Ana', 'Catalina','Eli']

df = pd.DataFrame(data, index=studenti)
sns.heatmap(df, annot=True, cmp='Ylgnbu')

plt.show()
```
