```python
class btn_primo_passo:
    import tkinter as tk

    def __init__(self, root, testo, comando):
        self._root = root
        self._testo = testo
        self._comando = comando
    
        self.btn = tk.Button(
            _root,
            text = _testo,
            command = _comando,          # funzione da chiamare al click
            bg="#2563eb",
            fg="black",
            font=("Arial", 12, "bold"),
            padx=20,
            pady=8,
            relief="flat",           # stile bordo: flat, raised, sunken
            cursor="hand2"          # cursore al passaggio del mouse
        )
    btn.pack(pady=60)

    def saluta():
        print("Bottone cliccato!")

    self.btn.bind("<Enter>")
    self.btn.bind("<Leave>")

    def on_enter(event):
        btn.configure(bg="#808080")

    root = tk.Tk()

    root.configure(bg="#0DA3A1")

    root.title("Bottoni in Tkinter")

    root.geometry("1000x1000")

    tk.Label().pack(pady=20)
    # padx/pady è il padding interno orizzontale/verticale
    # width/height sono le dimensioni in caratteri/pixel
    # label = widget non modificabile (è una costante)


    # Imposto la label 
    tk.Label(
        root,
        text ="Ciao, Tkinter!",
        font = ("Arial", 30, "italic"),
        bg = "#77a9ef",
        fg = "#000000",# colore del testo
        cursor = "hand2" 
    ).pack(pady=50)

     root.mainloop()
```
